```python
## read in codebook
df_pilot = pd.read_csv(r'/mnt/tscc/wed009/WholeGenome/data/FromBlair/pilot_encoding1.csv')
df_pilot
```

```python
df_adt = pd.read_csv(r'/mnt/tscc/wed009/WholeGenome/data/FromBlair/Adapters.csv',header=None)
names_adt = list(df_adt[0].str.split('_i').str.get(0))
seqs_adt = [lt.seqrc(sq[:20])for sq in df_adt[1]]
```

## adapter dic

```python
df_pilot_codebook = np.array(df_pilot.iloc[:, 7:].values,dtype=int)
nzRs = np.where(np.sum(df_pilot_codebook,axis=0)>0)[0]
is_combinatorial = np.zeros(1280)
is_combinatorial[:160]=1
is_combinatorial[160:(160+480)]=0 # 1 if comb, 0 if seq
is_combinatorial[(160+480):]=1
names_loc = list(df_pilot['locus'])
dic_readout_seqs = dict(zip(nzRs,zip(names_adt,seqs_adt,is_combinatorial[nzRs])))
```

```python
# re-index with the pilot bit
# 0-99 will be comb, matching 'bit' in the codebook
# 100- 159 will be seq
dic_pilot_idx = {}
icomb = 0
iseq = 100
for idx_full in dic_readout_seqs.keys():
    is_comb = dic_readout_seqs[idx_full][2]
    if is_comb == 1:
        dic_pilot_idx[idx_full] = icomb
        icomb += 1
    else:
        dic_pilot_idx[idx_full] = iseq
        iseq +=1
```

```python
dic_pilot_idx
```

```python
# create a new dictionary with the pilot index
dic_readout_seqs_pilot = dict(zip(dic_pilot_idx.values(), dic_readout_seqs.values()))
```

```python
dic_readout_seqs_pilot
```

```python
with open(os.path.join(save_folder, 'dic_readout_seqs_v1_pilot.pkl'), 'wb') as file:
    # Pickle the dictionary using the highest protocol available.
    pickle.dump(dic_readout_seqs_pilot, file, pickle.HIGHEST_PROTOCOL)
```

## V1 region dict

```python
dic_code_final = {}
iseq_start = 100
for ielem in range(len(names_loc)):
    nm_loc = names_loc[ielem]
    cd_loc = np.where(df_pilot_codebook[ielem])[0]
    cd_loc_mapped = np.vectorize(dic_pilot_idx.get)(cd_loc) # map it to codebook index
    
    dic_code_final[nm_loc] = [cd_loc_mapped[cd_loc_mapped >= iseq_start], # the seq bit
                             cd_loc_mapped[cd_loc_mapped < iseq_start]] # the 4 comb bits
```

# seq_codebook
6797 regions x 160 bits (100 comb + 60 seq)
binarized

```python
# create a 2D array and convert the on bit to 1
seq_codebook = np.zeros((len(dic_code_v1_pilot), len(dic_readout_seqs_pilot)))
print(seq_codebook.shape)               
for iR, region in enumerate(dic_code_v1_pilot.keys()):
    on_bits = np.concatenate((dic_code_v1_pilot[region][0], dic_code_v1_pilot[region][1]))
    seq_codebook[iR, on_bits] = 1
```

```python
region_names = list(dic_code_v1_pilot.keys())
bit_names = ['bit' + str(x) for x in np.arange(1, len(dic_readout_seqs_pilot) + 1)]
df_codebook_seq = pd.DataFrame(seq_codebook, index= region_names, columns= bit_names)
```
